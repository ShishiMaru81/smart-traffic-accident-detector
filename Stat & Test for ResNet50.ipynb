```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.metrics import classification_report, confusion_matrix
import numpy as np
```

```python
test_dir = "data/Test"
img_size = (224, 224)
batch_size = 32

test_datagen = ImageDataGenerator(
    preprocessing_function=tf.keras.applications.resnet50.preprocess_input
)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='binary',
    shuffle=False
)
```

```python
model = tf.keras.models.load_model("D:/Real Time Accident Detection/Models/ResNet50/accident_model_resnet50.h5")
```

```python
pred_probs = model.predict(test_generator)
preds = (pred_probs > 0.45).astype(int).flatten()


true_labels = test_generator.classes
class_names = list(test_generator.class_indices.keys())
```

```python
print("Class Names:", class_names)
print("\nConfusion Matrix:")
print(confusion_matrix(true_labels, preds))

print("\nClassification Report:")
print(classification_report(true_labels, preds, target_names=class_names))
```
